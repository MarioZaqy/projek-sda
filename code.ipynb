```python
from collections import deque

# CLASS GRAPH

class Graph:

    # Berfungsi sebagai constructor untuk membuat graph baru
    # self.graph digunakan untuk menyimpan data graph
    # self.directed digunakan untuk menentukan apakah graph berarah atau tidak
    def __init__(self, directed=False):
        self.graph = {}
        self.directed = directed

    # METHOD ADD_EDGE
    # Berfungsi untuk menambahkan hubungan/jalur antar node
    # u = node asal
    # v = node tujuan
    # jarak = bobot atau jarak antar node
    def add_edge(self, u, v, jarak):

        # Berfungsi untuk membuat node u jika belum ada
        if u not in self.graph:
            self.graph[u] = []

        # Berfungsi untuk membuat node v jika belum ada
        if v not in self.graph:
            self.graph[v] = []

        # Berfungsi untuk menambahkan hubungan dari u ke v
        self.graph[u].append((v, jarak))

        # Berfungsi untuk membuat graph 2 arah jika graph tidak berarah
        if not self.directed:
            self.graph[v].append((u, jarak))

    # BFS BIASA

    # Berfungsi untuk mencari node tujuan menggunakan algoritma BFS
    # BFS bekerja dengan sistem antrean (Queue)
    # BFS menelusuri graph secara melebar/per level
    def bfs(self, start, target):

        # Berfungsi untuk menyimpan node yang sudah dikunjungi
        visited = set()

        # Berfungsi sebagai queue/antrian BFS
        # Menyimpan:
        # (node sekarang, jalur/path, total jarak)
        queue = deque([(start, [start], 0)])

        # Berfungsi untuk menandai node awal sudah dikunjungi
        visited.add(start)

        # Berfungsi untuk menjalankan BFS selama queue masih ada isi
        while queue:

            # Berfungsi mengambil node paling depan pada queue
            (vertex, path, total_dist) = queue.popleft()

            # Berfungsi mengecek apakah node tujuan sudah ditemukan
            if vertex == target:
                return path, total_dist

            # Berfungsi menelusuri seluruh tetangga node
            for neighbour, weight in sorted(self.graph.get(vertex, [])):

                # Berfungsi agar node tidak dikunjungi berulang kali
                if neighbour not in visited:

                    # Menandai node sudah dikunjungi
                    visited.add(neighbour)

                    # Berfungsi menambahkan node baru ke queue
                    # Sekaligus menyimpan jalur dan total jarak
                    queue.append(
                        (neighbour,
                         path + [neighbour],
                         total_dist + weight)
                    )

        # Berfungsi mengembalikan nilai jika target tidak ditemukan
        return None, 0


    # DFS BIASA

    # Berfungsi untuk mencari node tujuan menggunakan algoritma DFS
    # DFS bekerja menggunakan Stack
    # DFS menelusuri graph sedalam mungkin terlebih dahulu
    def dfs(self, start, target):

        # Berfungsi menyimpan node yang sudah dikunjungi
        visited = set()

        # Berfungsi sebagai stack DFS
        # Menyimpan:
        # (node sekarang, jalur/path, total jarak)
        stack = [(start, [start], 0)]

        # Berfungsi menjalankan DFS selama stack masih ada isi
        while stack:

            # Berfungsi mengambil data terakhir dari stack
            (vertex, path, total_dist) = stack.pop()

            # Berfungsi mengecek apakah target ditemukan
            if vertex == target:
                return path, total_dist

            # Berfungsi mengecek apakah node belum dikunjungi
            if vertex not in visited:

                # Menandai node sudah dikunjungi
                visited.add(vertex)

                # Berfungsi menelusuri tetangga node
                # reversed digunakan agar urutan DFS tetap benar
                for neighbour, weight in reversed(sorted(self.graph.get(vertex, []))):

                    # Berfungsi agar node tidak dikunjungi berulang
                    if neighbour not in visited:

                        # Berfungsi menambahkan node baru ke stack
                        stack.append(
                            (neighbour,
                             path + [neighbour],
                             total_dist + weight)
                        )

        # Berfungsi mengembalikan nilai jika target tidak ditemukan
        return None, 0

    # BFS MODIFIKASI

    # Berfungsi mencari SEMUA kemungkinan jalur menggunakan BFS
    # Kemudian memilih jalur dengan total jarak terkecil
    def bfs_modified(self, start, target):

        # Berfungsi sebagai queue BFS
        queue = deque([(start, [start], 0)])

        # Berfungsi menyimpan semua kemungkinan jalur
        all_possible_paths = []

        # Berfungsi menjalankan BFS selama queue masih ada isi
        while queue:

            # Mengambil node paling depan dari queue
            (vertex, path, total_dist) = queue.popleft()

            # Berfungsi mengecek apakah target ditemukan
            if vertex == target:

                # Menyimpan jalur dan total jarak ke list
                all_possible_paths.append((path, total_dist))

                # Continue digunakan agar BFS tetap mencari jalur lain
                continue

            # Berfungsi menelusuri tetangga node
            for neighbour, weight in sorted(self.graph.get(vertex, [])):

                # Berfungsi mencegah node dikunjungi berulang
                # sekaligus mencegah infinite loop
                if neighbour not in path:

                    # Menambahkan node baru ke queue
                    queue.append(
                        (neighbour,
                        path + [neighbour],
                        total_dist + weight)
                    )

        # Berfungsi mengecek apakah ada jalur yang ditemukan
        if all_possible_paths:

            # Berfungsi mengurutkan jalur berdasarkan jarak terkecil
            all_possible_paths.sort(key=lambda x: x[1])

            # Mengembalikan jalur terbaik
            return all_possible_paths[0]

        # Berfungsi mengembalikan nilai jika jalur tidak ditemukan
        return None, 0


    # DFS MODIFIKASI

    # Berfungsi mencari SEMUA kemungkinan jalur menggunakan DFS
    # Kemudian memilih jalur dengan total jarak terkecil
    def dfs_modified(self, start, target):

        # Berfungsi sebagai stack DFS
        stack = [(start, [start], 0)]

        # Berfungsi menyimpan seluruh kemungkinan jalur
        all_possible_paths = []

        # Berfungsi menjalankan DFS selama stack masih ada isi
        while stack:

            # Mengambil data terakhir dari stack
            (vertex, path, total_dist) = stack.pop()

            # Berfungsi mengecek apakah target ditemukan
            if vertex == target:

                # Menyimpan jalur dan total jarak
                all_possible_paths.append((path, total_dist))

                # Continue digunakan agar DFS tetap mencari jalur lain
                continue

            # Berfungsi mengecek apakah node ada di graph
            if vertex in self.graph:

                # Menelusuri seluruh tetangga node
                for neighbour, weight in reversed(sorted(self.graph[vertex])):

                    # Berfungsi mencegah pengulangan node
                    # dan menghindari infinite loop
                    if neighbour not in path:

                        # Menambahkan node baru ke stack
                        stack.append(
                            (neighbour,
                            path + [neighbour],
                            total_dist + weight)
                        )

        # Berfungsi mengecek apakah ada jalur yang ditemukan
        if all_possible_paths:

            # Mengurutkan jalur berdasarkan jarak terkecil
            all_possible_paths.sort(key=lambda x: x[1])

            # Mengembalikan jalur terbaik
            return all_possible_paths[0]

        # Berfungsi mengembalikan nilai jika jalur tidak ditemukan
        return None, 0
```

```python
import time

# ─────────────────────────────────────────────
# FUNGSI PEMBANTU
# ─────────────────────────────────────────────

# Berfungsi untuk membangun graph dari data shelter area Padang (dari artikel referensi)
def build_graph_padang():
    g = Graph()
    g.add_edge('A', 'B', 1.5)
    g.add_edge('A', 'D', 1.5)
    g.add_edge('A', 'F', 3.0)
    g.add_edge('B', 'C', 1.5)
    g.add_edge('B', 'D', 1.8)
    g.add_edge('D', 'C', 1.1)
    g.add_edge('D', 'F', 0.8)
    g.add_edge('C', 'F', 2.5)
    g.add_edge('C', 'E', 0.2)
    g.add_edge('F', 'E', 1.5)
    g.add_edge('F', 'G', 1.0)
    return g


# Berfungsi untuk mengukur waktu rata-rata eksekusi sebuah fungsi
# Dijalankan sebanyak ULANGAN kali agar hasil lebih akurat
def ukur_waktu(fn, start, end, ulangan=10_000):
    mulai = time.perf_counter()
    for _ in range(ulangan):
        fn(start, end)
    selesai = time.perf_counter()
    # Berfungsi mengembalikan waktu rata-rata dalam mikrodetik (µs)
    return (selesai - mulai) / ulangan * 1_000_000


# Berfungsi untuk menampilkan hasil rute dan perbandingan waktu semua algoritma
def tampilkan_hasil(g, start, end):
    ULANGAN = 10_000

    print("-" * 56, flush=True)
    print(f"  Rute dari Shelter {start} ke Shelter {end}", flush=True)
    print("-" * 56, flush=True)

    # ── BFS Biasa ──────────────────────────────────────
    rute, jarak = g.bfs(start, end)
    waktu_bfs = ukur_waktu(g.bfs, start, end, ULANGAN)
    if rute:
        print(f"  BFS         : {' -> '.join(rute)}", flush=True)
        print(f"                Jarak : {jarak:.1f} KM".replace(".", ","), flush=True)
        print(f"                Waktu : {waktu_bfs:.4f} µs (rata-rata {ULANGAN:,}x)", flush=True)
    else:
        print("  BFS         : Rute tidak ditemukan", flush=True)
    print("-" * 56, flush=True)

    # ── DFS Biasa ──────────────────────────────────────
    rute, jarak = g.dfs(start, end)
    waktu_dfs = ukur_waktu(g.dfs, start, end, ULANGAN)
    if rute:
        print(f"  DFS         : {' -> '.join(rute)}", flush=True)
        print(f"                Jarak : {jarak:.1f} KM".replace(".", ","), flush=True)
        print(f"                Waktu : {waktu_dfs:.4f} µs (rata-rata {ULANGAN:,}x)", flush=True)
    else:
        print("  DFS         : Rute tidak ditemukan", flush=True)
    print("-" * 56, flush=True)

    # ── BFS Optimal ────────────────────────────────────
    rute, jarak = g.bfs_modified(start, end)
    waktu_bfs_opt = ukur_waktu(g.bfs_modified, start, end, ULANGAN)
    if rute:
        print(f"  BFS Optimal : {' -> '.join(rute)}", flush=True)
        print(f"                Jarak : {jarak:.1f} KM".replace(".", ","), flush=True)
        print(f"                Waktu : {waktu_bfs_opt:.4f} µs (rata-rata {ULANGAN:,}x)", flush=True)
    else:
        print("  BFS Optimal : Rute tidak ditemukan", flush=True)
    print("-" * 56, flush=True)

    # ── DFS Optimal ────────────────────────────────────
    rute, jarak = g.dfs_modified(start, end)
    waktu_dfs_opt = ukur_waktu(g.dfs_modified, start, end, ULANGAN)
    if rute:
        print(f"  DFS Optimal : {' -> '.join(rute)}", flush=True)
        print(f"                Jarak : {jarak:.1f} KM".replace(".", ","), flush=True)
        print(f"                Waktu : {waktu_dfs_opt:.4f} µs (rata-rata {ULANGAN:,}x)", flush=True)
    else:
        print("  DFS Optimal : Rute tidak ditemukan", flush=True)
    print("-" * 56, flush=True)

    # ── Ringkasan Perbandingan Waktu ───────────────────
    print(flush=True)
    print("  PERBANDINGAN WAKTU EKSEKUSI", flush=True)
    print("  " + "=" * 44, flush=True)

    # Berfungsi untuk mengumpulkan dan mengurutkan hasil waktu
    hasil_waktu = [
        ("BFS",         waktu_bfs),
        ("DFS",         waktu_dfs),
        ("BFS Optimal", waktu_bfs_opt),
        ("DFS Optimal", waktu_dfs_opt),
    ]
    # Berfungsi untuk mengurutkan dari yang tercepat ke terlambat
    hasil_waktu.sort(key=lambda x: x[1])
    waktu_tercepat = hasil_waktu[0][1]

    for peringkat, (nama, waktu) in enumerate(hasil_waktu, 1):
        # Berfungsi untuk menghitung seberapa lambat dibanding yang tercepat
        rasio = waktu / waktu_tercepat
        label_peringkat = "🥇" if peringkat == 1 else f"#{peringkat}"
        print(f"  {label_peringkat}  {nama:<12} : {waktu:>8.4f} µs  (x{rasio:.2f})", flush=True)

    print("  " + "=" * 44, flush=True)
    print(f"  Tercepat : {hasil_waktu[0][0]}", flush=True)
    print(f"  Terlambat: {hasil_waktu[-1][0]}", flush=True)
    print(flush=True)


# ─────────────────────────────────────────────
# MENU UTAMA
# ─────────────────────────────────────────────

print("=" * 56, flush=True)
print("   SISTEM PENCARIAN RUTE EVAKUASI BENCANA", flush=True)
print("   Algoritma BFS & DFS — Struktur Data", flush=True)
print("=" * 56, flush=True)
print(flush=True)
print("  Pilih mode penggunaan:", flush=True)
print("  1. Gunakan data area Padang (Shelter A-G)", flush=True)
print("  2. Input graf sendiri (Graf Kustom)", flush=True)
print(flush=True)

# Berfungsi untuk membaca pilihan mode dari pengguna
pilihan = input("  Masukkan pilihan (1/2): ").strip()
print(f"  ✓ Pilihan: {pilihan}", flush=True)
print(flush=True)


# ─────────────────────────────────────────────
# MODE 1 — DATA AREA PADANG (dari artikel referensi)
# ─────────────────────────────────────────────

if pilihan == "1":
    print("=" * 56, flush=True)
    print("   MODE 1: DATA AREA PADANG", flush=True)
    print("=" * 56, flush=True)
    print(flush=True)

    # Berfungsi untuk mendefinisikan daftar shelter yang tersedia
    SHELTERS = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    print(f"  Shelter tersedia: {', '.join(SHELTERS)}", flush=True)
    print(flush=True)

    # Berfungsi untuk meminta shelter awal dari pengguna
    start = input("  Masukkan shelter awal  : ").strip().upper()
    print(f"  ✓ Shelter awal  : {start}", flush=True)

    # Berfungsi untuk meminta shelter tujuan dari pengguna
    end = input("  Masukkan shelter tujuan: ").strip().upper()
    print(f"  ✓ Shelter tujuan: {end}", flush=True)
    print(flush=True)

    # Berfungsi untuk memvalidasi input pengguna
    if start not in SHELTERS:
        print(f"  Shelter '{start}' tidak valid. Pilih dari: {', '.join(SHELTERS)}", flush=True)
    elif end not in SHELTERS:
        print(f"  Shelter '{end}' tidak valid. Pilih dari: {', '.join(SHELTERS)}", flush=True)
    elif start == end:
        print("  Shelter awal dan tujuan tidak boleh sama.", flush=True)
    else:
        # Berfungsi untuk membangun graph dan menjalankan semua algoritma
        g = build_graph_padang()
        tampilkan_hasil(g, start, end)


# ─────────────────────────────────────────────
# MODE 2 — GRAF KUSTOM (input pengguna sendiri)
# ─────────────────────────────────────────────

elif pilihan == "2":
    print("=" * 56, flush=True)
    print("   MODE 2: GRAF KUSTOM", flush=True)
    print("=" * 56, flush=True)
    print(flush=True)

    # Berfungsi untuk meminta jumlah shelter dari pengguna
    while True:
        try:
            n_shelter = int(input("  Masukkan jumlah shelter: ").strip())
            if n_shelter < 2:
                print("  Minimal 2 shelter diperlukan.", flush=True)
                continue
            break
        except ValueError:
            print("  Masukkan angka yang valid.", flush=True)

    # Berfungsi untuk meminta nama setiap shelter dari pengguna
    print(flush=True)
    shelters_kustom = []
    for i in range(n_shelter):
        while True:
            nama = input(f"  Nama shelter ke-{i+1}: ").strip()
            if not nama:
                print("  Nama shelter tidak boleh kosong.", flush=True)
                continue
            if nama in shelters_kustom:
                print(f"  Shelter '{nama}' sudah ada, gunakan nama lain.", flush=True)
                continue
            shelters_kustom.append(nama)
            print(f"  ✓ Shelter ke-{i+1}: {nama}", flush=True)
            break

    # Berfungsi untuk meminta jumlah koneksi (edge) antar shelter
    print(flush=True)
    while True:
        try:
            n_koneksi = int(input("  Masukkan jumlah koneksi antar shelter: ").strip())
            if n_koneksi < 1:
                print("  Minimal 1 koneksi diperlukan.", flush=True)
                continue
            break
        except ValueError:
            print("  Masukkan angka yang valid.", flush=True)

    # Berfungsi untuk membangun graph kustom dari input pengguna
    g_kustom = Graph()

    for i in range(n_koneksi):
        print(f"\n  --- Koneksi ke-{i+1} ---", flush=True)

        # Berfungsi untuk meminta shelter asal koneksi
        while True:
            frm = input(f"  Dari shelter ({', '.join(shelters_kustom)}): ").strip()
            if frm not in shelters_kustom:
                print(f"  Shelter '{frm}' tidak ada. Pilih dari daftar.", flush=True)
                continue
            break
        print(f"  ✓ Dari : {frm}", flush=True)

        # Berfungsi untuk meminta shelter tujuan koneksi
        while True:
            to = input(f"  Ke shelter   ({', '.join(shelters_kustom)}): ").strip()
            if to not in shelters_kustom:
                print(f"  Shelter '{to}' tidak ada. Pilih dari daftar.", flush=True)
                continue
            if to == frm:
                print("  Tidak boleh sama dengan shelter asal.", flush=True)
                continue
            break
        print(f"  ✓ Ke   : {to}", flush=True)

        # Berfungsi untuk meminta jarak antar dua shelter
        while True:
            try:
                jarak = float(input("  Jarak (km)              : ").strip())
                if jarak <= 0:
                    print("  Jarak harus lebih dari 0.", flush=True)
                    continue
                break
            except ValueError:
                print("  Masukkan angka yang valid.", flush=True)
        print(f"  ✓ Jarak: {jarak} km", flush=True)

        # Berfungsi untuk menambahkan koneksi ke dalam graph kustom
        g_kustom.add_edge(frm, to, jarak)

    # Berfungsi untuk menampilkan shelter tersedia dan meminta rute pencarian
    print(flush=True)
    print(f"  Shelter tersedia: {', '.join(shelters_kustom)}", flush=True)
    print(flush=True)

    # Berfungsi untuk meminta shelter awal dari pengguna
    while True:
        start = input("  Masukkan shelter awal  : ").strip()
        if start not in shelters_kustom:
            print(f"  Shelter '{start}' tidak ada dalam daftar.", flush=True)
            continue
        break
    print(f"  ✓ Shelter awal  : {start}", flush=True)

    # Berfungsi untuk meminta shelter tujuan dari pengguna
    while True:
        end = input("  Masukkan shelter tujuan: ").strip()
        if end not in shelters_kustom:
            print(f"  Shelter '{end}' tidak ada dalam daftar.", flush=True)
            continue
        if end == start:
            print("  Shelter awal dan tujuan tidak boleh sama.", flush=True)
            continue
        break
    print(f"  ✓ Shelter tujuan: {end}", flush=True)
    print(flush=True)

    # Berfungsi untuk menjalankan semua algoritma dan menampilkan hasil + waktu
    tampilkan_hasil(g_kustom, start, end)


# ─────────────────────────────────────────────
# INPUT TIDAK VALID
# ─────────────────────────────────────────────

else:
    print("  Pilihan tidak valid. Masukkan 1 atau 2.", flush=True)
```
